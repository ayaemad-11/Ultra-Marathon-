```python
new_columns = {
    'Year of event': 'year',
    'Event dates': 'event_dates',
    'Event name': 'event_name',
    'Event distance/length': 'event_distance_length',
    'Event number of finishers': 'event_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_avg_speed',
    'Athlete ID': 'athlete_id',
    
}

df2.rename(columns=new_columns, inplace=True)
```

```python
# graphs to visulization 
sns.histplot( data = df2, x = 'event_distance_length', hue = 'athlete_gender')
```

```python
#50mi
import seaborn as sns
sns.displot(df2[df2['event_distance_length']=='50mi']['athlete_avg_speed'],color = "purple" , alpha = 1)
```

```python
sns.violinplot(data = df2, x ='event_distance_length', y='athlete_avg_speed', hue='athlete_gender', split=True , inner='quart', linewidth=1)
```

```python
df2.groupby(['event_distance_length','athlete_gender'])['athlete_avg_speed'].mean()
```
